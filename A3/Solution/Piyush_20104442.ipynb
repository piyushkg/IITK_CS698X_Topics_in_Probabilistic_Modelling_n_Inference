```python
import numpy as np
import matplotlib.pyplot as plt
```

# Rejection Sampling

```python
sig = 3
M=40
s_select = np.zeros(10000)
j=0
while j < 10000:
    s = np.random.normal(0, sig, 1)
    if s > -np.pi and s < np.pi:
        p_tilde = np.exp(np.sin(s))
        q = (1/np.sqrt(2*np.pi*(sig**2)))*np.exp((-0.5/(sig**2))*s*s)
        u = np.random.uniform(0,M*q,1)
        if u <= p_tilde:
             s_select[j] = s
             j = j+1   
```

```python
s_select.shape
```

```python
plot1 = plt.figure(1)
count, bins, ignored = plt.hist(s_select, 200)
plt.xlabel('x-data', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.suptitle('Rejection Sampling Histogram for exp(sin(x))', fontsize=20)
plt.grid(True, color = "grey", linewidth = "1", linestyle = "-.")
plt.show()
```
